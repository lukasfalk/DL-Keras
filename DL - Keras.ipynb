Loading the datase into a pandas dataframe

```python
import pandas as pd
import numpy as np

# Loading the dataset directly from the link
concrete_data = pd.read_csv('https://cocl.us/concrete_data')
concrete_data.shape
concrete_data.head()
```

# A. Build a baseline model

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Defining X and y
concrete_data_columns = concrete_data.columns
X = concrete_data.loc[:, concrete_data.columns != 'Strength']
y = concrete_data['Strength']

# Defining the function for creating the model with one hidden layer
def one_layer_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(X.shape[1],)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

baseline_model = one_layer_model()
iterations = 50
MSEs = []
# Repeating steps 1-3 50 times
for i in range(1, iterations+1):
    print("\nIteration {}/{}".format(i, iterations))
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=69)
    baseline_model.fit(X_train, y_train, validation_split=0.3, epochs=50, verbose=1)
    y_predict = baseline_model.predict(X_test)
    MSEs.append(mean_squared_error(y_test, y_predict))

print("Mean: {}".format(np.mean(MSEs)))
print("STD: {}".format(np.std(MSEs)))
```

# B. Normalize the data

```python
# Normalizing the data
X_norm = (X - X.mean()) / X.std()
y_norm = (y - y.mean()) / y.std()

norm_MSEs = []
# Repeating steps 1-3 50 times again, but with normalized data
for i in range(1, iterations+1):
    print("\nIteration {}/{}".format(i, iterations))
    X_train, X_test, y_train, y_test = train_test_split(X_norm, y_norm, test_size=0.3, random_state=69)
    baseline_model.fit(X_train, y_train, validation_split=0.3, epochs=50, verbose=1)
    y_predict = baseline_model.predict(X_test)
    norm_MSEs.append(mean_squared_error(y_test, y_predict))

print("Step A Mean: {}".format(np.mean(MSEs)))
print("Step B Mean: {}".format(np.mean(norm_MSEs)))
```

As we can see, the mean and STD of all the mean square errors are much closer to 0 than in part A

# C. Increase the number of epochs

```python
epoch_MSEs = []
# Repeating steps 1-3 50 times again, but with normalized data
for i in range(1, iterations+1):
    print("\nIteration {}/{}".format(i, iterations))
    X_train, X_test, y_train, y_test = train_test_split(X_norm, y_norm, test_size=0.3, random_state=69)
    baseline_model.fit(X_train, y_train, validation_split=0.3, epochs=100, verbose=1)
    y_predict = baseline_model.predict(X_test)
    epoch_MSEs.append(mean_squared_error(y_test, y_predict))

print("Step C Mean: {}".format(np.mean(epoch_MSEs)))
print("Step B Mean: {}".format(np.mean(norm_MSEs)))
```

The mean decreases when the epochs increase

# D. Increase the number of hidden layers

```python
# Defining the function for creating the model with three hidden layer
def three_layer_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(X.shape[1],)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

baseline_model_3 = three_layer_model()

layers_MSEs = []
# Repeating steps 1-3 50 times again, but with normalized data
for i in range(1, iterations+1):
    print("\nIteration {}/{}".format(i, iterations))
    X_train, X_test, y_train, y_test = train_test_split(X_norm, y_norm, test_size=0.3, random_state=69)
    baseline_model_3.fit(X_train, y_train, validation_split=0.3, epochs=100, verbose=1)
    y_predict = baseline_model_3.predict(X_test)
    layers_MSEs.append(mean_squared_error(y_test, y_predict))

print("Step D Mean: {}".format(np.mean(layers_MSEs)))
print("Step B Mean: {}".format(np.mean(norm_MSEs)))
```

The increase in hidden layers has increased the mean
